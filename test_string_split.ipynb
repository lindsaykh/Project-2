```python
import requests
import pandas as pd
import numpy as np
from pprint import pprint
# import config
```

```python
# Import json and convert to DF
url = 'https://www.thecocktaildb.com/api/json/v2/9973533/search.php?s'

response = requests.get(url).json()

```

```python
drinks_df = pd.json_normalize(response['drinks'], max_level=1)
drinks_df.head()
```

```python
drinks_df.columns
```

```python
# columns to new df for instructions, ingredients, measures

measure_df = drinks_df[['idDrink', 'strDrink',  'strIngredient1', 'strIngredient2', 'strIngredient3',
       'strIngredient4', 'strIngredient5', 'strIngredient6', 'strIngredient7',
       'strIngredient8', 'strMeasure1', 'strMeasure2',
       'strMeasure3', 'strMeasure4', 'strMeasure5', 'strMeasure6',
       'strMeasure7', 'strMeasure8']]

measure_df.info()
```

```python
# splitting up the strings in the measures columns 
measure_df['strMeasure1']
```

```python
measurements_df = measure_df['strMeasure1'].str.split(' ', 2, expand=True)
# measurements_df.to_excel("measurements.xlsx")
measurements_df.head(20)
```

```python
measurements = []
units = []
# check if numeric or alpha 
for index, row in measure_df.head().iterrows():
    
    string_list = row['strMeasure1'].split(' ', 1)
    
    for x in string_list:
        
        if x.isnumeric() == True:
    
            measurements.append(x)
    
        else:
            
            units.append(x)
    
    measurements_df = pd.DataFrame(
        {"Measurements": measurements}
#         {"Units": units}
    )
    
#     units_df = pd.DataFrame({"Units": units})
        
measurements_df
```

```python
# use the apply function to do the split 

def split(row):
    
    # in this column, split by " " and take the last item of list
    
    row['strMeasure1'] = row['strMeasure1'].str.split(" ")[-1]
    
    return row 

test = measure_df.apply(split)
test
```
