```python
import pandas as pd
```

```python
drink_data = "./cocktail_data.csv"
```

```python
drink_data_df = pd.read_csv(drink_data)
```

```python
drink_data_df.head()
```

```python
drink_data_abv = drink_data_df[["idDrink", "strDrink", "Drink_ABV"]]
drink_data_abv.head()
```

```python
import requests

# Import json and convert to DF
url = 'https://www.thecocktaildb.com/api/json/v2/9973533/search.php?s'

response = requests.get(url).json()

drinks_df = pd.json_normalize(response['drinks'], max_level=1)
drinks_df.head(100)
```

```python
drinks_df.head(100)
```

```python
# change idDrink data type for merge
drinks_df[['idDrink']] = drinks_df[['idDrink']].apply(pd.to_numeric)
```

```python
drinks = pd.merge(drink_data_abv, drinks_df, on='idDrink', how='left')
drinks.head()
```

```python
drinks_bac = drinks[["idDrink","strDrink_x","Drink_ABV", "strImageSource"]]
```

```python
drinks_bac = drinks_bac.rename(columns={'strDrink_x': 'strDrink', 'strImageSource': 'image'})
```

```python
drinks_bac.info()
```

```python
#add dummy pic for empty image cells
drinks_bac["image"].fillna("https://www.rokaakor.com/wp-content/uploads/2018/10/Top-5-Cocktail-Trends.jpg", inplace = True)
```

```python
drinks_bac.head()
```

```python
drinks_bac.to_csv(r'.\drinks_bac.csv')
```
